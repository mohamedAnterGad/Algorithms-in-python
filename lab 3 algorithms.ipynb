## imports

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.Graph()
graph.add_nodes_from([(0,{'color':'green'}),1,2,3,4,5,6,7,8])
graph.add_edges_from([(0,3,{'weight': 3.1415}),(0,8,{'weight': 3.1415}),(0,1,{'weight': 3.1415}),(1,7,{'weight': 3.1415}),(7,2,{'weight': 3.1415}),(2,3,{'weight': 3.1415}),(2,5,{'weight': 3.1415}),(5,6,{'weight': 3.1415}),(3,4,{'weight': 3.1415}),(4,8,{'weight': 3.1415})])
nx.draw(graph,with_labels = True, font_weight='bold')
plt.show()
```

### DFS path

```python
#print(list(nx.edge_dfs(graph,source= 0))) #generate all the edges of the graph
print(list(nx.dfs_edges(graph,source = 0)))
```

### BFS path

```python
print(list(nx.bfs_edges(graph,source=0)))
```

## second graph

```python
G = nx.Graph()
G.add_weighted_edges_from([(0, 1, 8), (0, 2, 9), (1, 3, 1), (1, 5, 7), \
                           (1, 6, 9), (2, 5, 6), (2, 6, 4), (3, 7, 4), \
                           (4, 6, 7), (4, 7, 7), (5, 7, 2), (6, 7, 6)])
pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels);
```

### Dijkstra algorithm

### Prim algorithm

```python
T = nx.minimum_spanning_tree(G,algorithm = 'prim')
sorted(T.edges(data = True))
```

### kruskal algorithm

```python
T = nx.minimum_spanning_tree(G,algorithm = 'kruskal')
sorted(T.edges(data = True))
```
